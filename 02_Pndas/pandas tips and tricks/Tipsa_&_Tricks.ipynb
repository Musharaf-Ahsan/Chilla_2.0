# Splitting a String into Multiple Columns
- All array must of same length.
- We'll create dataset using dictionary method.
- We can split a string into multiple columns using this command **"df.column_name.str.split(' ').head()**.

```python
# import Libraries
import pandas as pd
df = pd.DataFrame({"name":["Muhammad Bilal","Hussnain Ali","Jwad Hassan","Muhammad Akif"],
                   "Location":["Bahawalpur, Pakistan","Lahore, Pakistan","Lodhran, Pakistan","Jhaniyan, Pakistan"],})
df
```

```python
# Splitting a column into two columns
df.name.str.split(" ").head()
```

```python

df.name.str.split(" ",expand=True)
```

```python
# Adding those splits into new columns
# Now make it more accurate.
df[["First_name", "Last_name"]]=df.name.str.split(" ",expand=True)
df.head()
```

```python
# Now split Location into two columns city and country
df[["City", "Country"]]=df.Location.str.split(", ",expand=True)
df.head()
```

```python
# Refine the data
df = df[['First_name','Last_name','City','Country']]
df.head()
```

# 23- Reshaping the DataFrame

```python
from turtle import distance


distance = pd.DataFrame([['12345', 100, 200, 300], ['34567', 400, 500, 600], ['67890', 700, 800, 900]],
                          columns=['zip', 'factory', 'warehouse', 'retail'])
distance.head()
```

```python
sns.barplot(x='zip', y='factory', data=distance)
```

```python
distance2 = pd.DataFrame([[1, '12345', 'factory'], [2, '34567', 'warehouse']],
                      columns=['user_id', 'zip', 'location_type'])
distance2.head()
```

#### Now I want to reshape "distance" DataFrame like "distance2" in long format

```python
# melt ka function use krty huye data ko long form me convert kr dia.
# Data ko is form me change krne se plot easy or achy bn jate hain by baba g.
distance_long = distance.melt(id_vars='zip', var_name='location_type', value_name='distance')
distance_long.head()
```

```python
distance_long.dtypes
```

```python
import seaborn as sns
sns.barplot(x='zip', y='distance', hue='location_type', data=distance_long)
```
